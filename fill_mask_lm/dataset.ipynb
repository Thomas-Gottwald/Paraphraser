```python
import pandas as pd
import seaborn as sns
```

```python
def plot_original_POS(df_dataset):
    df = df_dataset.reset_index().groupby('og_POS').agg({'chosen': 'sum'}).rename(columns={'chosen': 'count'})
    df.plot.bar()

def plot_suggested_POS(df_dataset):
    df = df_dataset.reset_index().groupby('POS').agg({'suggested': 'sum'})
    df.plot.bar()

def plot_chosen_POS(df_dataset):
    df = df_dataset.reset_index().groupby('POS').agg({'chosen': 'sum'})
    df.plot.bar()

def plot_POS_dependenc(df_dataset, POS_tag, chosen=False):
    kind = 'suggested'
    if chosen:
        kind = 'chosen'
    df_dataset.query("og_POS == '{}'".format(POS_tag))[[kind]].plot.bar()

def plot_POS_heatmap(df_dataset, chosen=False):
    kind = 'suggested'
    if chosen:
        kind = 'chosen'
    df = df_dataset[[kind]].unstack(level=-1)
    df.div(df.sum(axis=1), axis=0)
    sns.heatmap(df.div(df.sum(axis=1), axis=0), cmap="viridis")
```

```python
def avg_score(df_dataset, chosen=False):
    kind = 'suggested'
    kind_score = 'avg_sug_score'
    if chosen:
        kind = 'chosen'
        kind_score = 'avg_cho_score'
    print((df_dataset[kind_score]*df_dataset[kind]).sum()/df_dataset[kind].sum())
```

# RoBERTa

# Wikipedia

```python
df_path = './../data/wikipedia/sp(roberta-large,0.5)/dataset.pkl'
df_dataset = pd.read_pickle(df_path)
df_dataset
```

```python
plot_original_POS(df_dataset)
```

```python
plot_suggested_POS(df_dataset)
```

```python
plot_chosen_POS(df_dataset)
```

```python
plot_POS_dependenc(df_dataset, 'NOUN', chosen=False)
```

```python
plot_POS_dependenc(df_dataset, 'NOUN', chosen=True)
```

```python
plot_POS_heatmap(df_dataset, chosen=False)
```

```python
plot_POS_heatmap(df_dataset, chosen=True)
```
